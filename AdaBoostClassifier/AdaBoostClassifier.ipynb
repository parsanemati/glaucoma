```python
data_selection = df.iloc[:,29:1054].fillna(0)
data_selection.shape
```

```python
# Splitting the data into 300 training instances and 104 test instances
n = 157
all_Ids = np.arange(len(data_selection))

# random.shuffle(all_Ids)
test_Ids = all_Ids[0:n]
train_Ids = all_Ids[n:]
data_test = data_selection.loc[test_Ids, :]
data_train = data_selection.loc[train_Ids, :]
# print(data_train.shape) , print(data_test.shape)

X_train = data_train.iloc[:,0:1024]
Y_train = data_train['Real Label']

x_test = data_test.iloc[:,0:1024]


y_test = data_test['Real Label']
print("done")
```
